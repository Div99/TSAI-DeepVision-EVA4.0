## Import Sodium's YOLO or Solo

```
import sodium.yolo as solo
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
yolo = solo.YoloOpenCV()
```

## Download the images

```
import urllib.request
print('Beginning file download with urllib2...')
BASE_URL = 'https://github.com/satyajitghana/TSAI-DeepVision-EVA4.0/blob/master/13_YoloV3/images/'
files = ['beagle.jpg', 'its_me.jpg', 'potted_plant.jpg']
[urllib.request.urlretrieve(BASE_URL + file + '?raw=true', file) for file in files]
```

```
img = yolo.run_yolo_on_image('beagle.jpg')
plt.figure(figsize=(10, 9))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
img = yolo.run_yolo_on_image('its_me.jpg')
plt.figure(figsize=(10, 15))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```
img = yolo.run_yolo_on_image('potted_plant.jpg')
plt.figure(figsize=(10, 15))
plt.imshow(img)
plt.axis('off')
plt.show()
```
